IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

IMPORTING DATASET

```python
dataset = pd.read_csv('Restaurant_Reviews.csv', delimiter = ',', quotechar = '"')
x = dataset.iloc[:, 0].values
y = dataset.iloc[:, 1].values

print(dataset.columns)
display(dataset.head())
```

CLEANING THE DATASET

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []

for i in range(0, 1000):
  reviews = re.sub('[^a-zA-Z]', ' ', dataset['Review'][i])
  reviews = reviews.lower()
  reviews = reviews.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  reviews = [ps.stem(word) for word in reviews if word not in (all_stopwords)]
  reviews = ' '.join(reviews)
  corpus.append(reviews)



for review in corpus:
    print(review, '\n')


```

CREATING THE BAG OF WORDS

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:, -1].values

print(len(x[0]))

print(x)
print('\n\n',y)
```

SPLITTING THE DATASET

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

MODELS TRAINING

TRAINING THE SVM(SUPPORT VECTOR MACHINE) ON TRAINING

```python
from sklearn.svm import SVC
classifier1 = SVC(kernel='linear', random_state=0)
classifier1.fit(X_train, y_train)
```

TRAINING THE RANDOM FOREST ON TRAINING SET

```python
from sklearn.ensemble import RandomForestClassifier
classifier2 = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
classifier2.fit(X_train, y_train)
```

PREDICTING THE TEST SET WITH SVM

```python
y_pred_1 = classifier1.predict(X_test)
print(np.concatenate((y_pred_1.reshape(len(y_pred_1),1), y_test.reshape(len(y_test),1)),1))
```

PREDICTING TEST SET WITH RANDOM FOREST

```python
y_pred_2 = classifier2.predict(X_test)
print(np.concatenate((y_pred_2.reshape(len(y_pred_2),1), y_test.reshape(len(y_test),1)),1))
```

CONFUSION MATRIX FOR SVM

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_test, y_pred_1)
print(cm)
accuracy_score(y_test, y_pred_1)
```

CONFUSTION METRIX FOR RANDOM FOREST

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_test, y_pred_2)
print(cm)
accuracy_score(y_test, y_pred_2)
```

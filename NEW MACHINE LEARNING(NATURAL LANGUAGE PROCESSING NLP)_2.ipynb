NATURAL LANGUAGE PROCESSING

IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

IMPORTING DATASETS

```python
dataset = pd.read_csv('Restaurant_Reviews.csv', quoting=3, on_bad_lines='skip')
x = dataset.iloc[:, 0].values
y = dataset.iloc[:, 1].values

print(dataset.columns)
display(dataset.head())
```

CLEANING THE TEXTS

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []

for i in range(0, len(dataset)):
  reviews = re.sub('[^a-zA-Z]', ' ', dataset['Review'][i])
  reviews = reviews.lower()
  reviews = reviews.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  reviews = [ps.stem(word) for word in reviews if word not in (all_stopwords)]
  reviews = ' '.join(reviews)
  corpus.append(reviews)



for review in corpus:
    print(review, '\n')
```

CREATING THE BAG OF WORDS MODEL

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:, -1].values

print(len(x[0]))

print(x)
print('\n\n',y)
```

SPLITTING THE DATASET INTO TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

TRAINING THE NAVIE BAYES MODEL ON THE TRAINING SET

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

PREDICTING THE TEST SET RESULT

```python
y_predict = classifier.predict(x_test)
print(np.concatenate((y_predict.reshape(len(y_predict), 1), y_test.reshape(len(y_test), 1)), 1))
```

MAKING THE CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_test, y_predict)
print(cm)
accuracy_score(y_test, y_predict)
```

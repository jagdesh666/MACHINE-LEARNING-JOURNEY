IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

IMPORTING DATASET

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

SPLITING DATASET INTO TRAINING SET AND TESTING SET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

TRAINING NAVIE BAYES MODEL ON TRAINING SET

```python
from sklearn.naive_bayes import GaussianNB
regressor = GaussianNB()
regressor.fit(x_train,y_train)
```

PREDICTING THE NEW RESULT

```python
age = 30
salary = 87000
regressor.predict(sc.transform([[age,salary]]))
```

PREDICTING THE TEST SET RESULT

```python
y_pred = regressor.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

MAKING THE CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

VISUALISING THE TRAINING SET

```python
from matplotlib.colors import ListedColormap
X1, X2 = np.meshgrid(np.arange(start = x_train[:, 0].min() - 1, stop = x_train[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_train[:, 1].min() - 1, stop = x_train[:, 1].max() + 1, step = 0.01))
```

```python
y_predict = regressor.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape)
```

```python
plt.contourf(X1, X2, y_predict, alpha = 0.75, cmap = ListedColormap(('red', 'green')), origin='lower')
```

```python
for i, j in enumerate(np.unique(y_train)):
    plt.scatter(x_train[y_train == j, 0], x_train[y_train == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('naive bayes (training set)')
plt.xlabel('age')
plt.ylabel('estimated salary')
plt.legend()
plt.show()
```

VISUALISING THE TESTING SET RESULT

```python
from matplotlib.colors import ListedColormap
import numpy as np
import matplotlib.pyplot as plt

# Create meshgrid based on x_test
X1, X2 = np.meshgrid(
    np.arange(start = x_test[:, 0].min() - 1, stop = x_test[:, 0].max() + 1, step = 0.01),
    np.arange(start = x_test[:, 1].min() - 1, stop = x_test[:, 1].max() + 1, step = 0.01)
)

y_predict = regressor.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape)
plt.contourf(X1, X2, y_predict, alpha = 0.75, cmap = ListedColormap(('red', 'green')), origin='lower')
for i, j in enumerate(np.unique(y_test)):
    plt.scatter(x_test[y_test == j, 0], x_test[y_test == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)

plt.title('naive bayes (test set)')
plt.xlabel('age')
plt.ylabel('estimated salary')
plt.legend()
plt.show()
```
